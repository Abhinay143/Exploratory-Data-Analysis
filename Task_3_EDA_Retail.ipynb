**Sub-Category Value Counts**

```
ax=df["Sub-Category"].value_counts().plot(kind="bar",figsize=(10,5))
ax.set_xlabel("Sub_Category")
ax.set_ylabel("Count")
plt.style.use("seaborn")
plt.show()
```

```
df["Sub-Category"].unique()
```

```
label=['Bookcases', 'Chairs', 'Labels', 'Tables', 'Storage',
       'Furnishings', 'Art', 'Phones', 'Binders', 'Appliances', 'Paper',
       'Accessories', 'Envelopes', 'Fasteners', 'Supplies', 'Machines',
       'Copiers']
plt.pie(df["Sub-Category"].value_counts(),labels=label,shadow=True,radius=2,autopct="%1.1f%%")
plt.show()
```

Visualizing Sales and Profit for each sub-category

```
sales_profit=df.groupby("Sub-Category")["Sales","Profit"].sum()
sales_profit.plot(kind="bar",width=0.8)
plt.style.use("seaborn")
plt.show()
```

```
print(plt.style.available)
```

**Sub-Categories Distribution in each region**

```
ax=sns.countplot(x="Sub-Category",data=df,hue="Region")
plt.title("Count of sub_categories distributed in each Region")
plt.xticks(rotation=90)
plt.show()
```

**Calculating CostPrice**

```
df["Cost_Price"]=df["Sales"]-df["Profit"]
df["Cost_Price"]
```

```
cp=df.groupby("Sub-Category")["Cost_Price"].sum()
cp.plot(kind="bar",color="#ff6633")
plt.ylabel("costprice")
plt.show()
```

**Region-Wise Sales**

```
region_sum=df.groupby("Region")["Sales"].sum()
region_sum.plot(kind="bar",color="#66ff66",width=0.5)
plt.ylabel("Sales")
plt.show()
```

**State-Wise sales**

```
city_sum=df.groupby("State")["Sales"].sum()
fig=plt.figure(figsize=(10,15))
city_sum.plot(kind="bar")
plt.show()
```
